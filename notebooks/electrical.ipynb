```python
import pandas as pd

df = pd.read_csv('electeric.csv')

df.head()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df['Price.USD'] = df['Price.USD'].str.replace(',', '')
df['Price.USD'] = df['Price.USD'].str.replace('$', '')
df['Price.USD'] = df['Price.USD'].astype(float)
```

```python
df.info()
```

```python
df['price'] = df['Price.USD']
df = df.drop(['Price.USD'], axis=1)
```

```python
df.head()
```

```python
df.Car_name.value_counts()
```

```python
df.drop(columns=['Car_name'], inplace=True)
```

```python
df.columns
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for i in df.columns:
    sns.histplot(df[i])
    plt.show()
```

```python
corr = df.corr()

hm = sns.heatmap(data=corr,
                annot=True)
```
